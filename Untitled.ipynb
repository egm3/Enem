```python
import matplotlib.pyplot as plt
import numpy as np

# Valores de xi e y observados
xi = [20, 20, 20, 20, 25, 25, 25, 25, 30, 30, 30, 30,
      35, 35, 35, 35, 40, 40, 40, 40]
y_observado = [96, 92, 106, 100, 98, 104, 110, 101,
               116, 106, 106, 109, 100, 112, 105, 118,
               108, 113, 112, 127]

# Calcula os valores de yi pela funo
yi_modelo = [80.5 + 0.9 * x for x in xi]

# Scatter dos valores observados
plt.scatter(xi, y_observado, color='blue', label='y observado', alpha=0.7)

# Plot da linha da funo yi
# Para uma linha suave, criaremos pontos nicos de xi ordenados
xi_unicos = np.array(sorted(set(xi)))
yi_linha = 80.5 + 0.9 * xi_unicos
plt.plot(xi_unicos, yi_linha, color='red', label='yi = 80.5 + 0.9*xi')

# Esttica do grfico
plt.xlabel('xi')
plt.ylabel('y / yi')
plt.title('Valores observados vs. Funo linear')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# Dados
xi = [20, 20, 20, 20, 25, 25, 25, 25, 30, 30, 30, 30,
      35, 35, 35, 35, 40, 40, 40, 40]
y_observado = [96, 92, 106, 100, 98, 104, 110, 101,
               116, 106, 106, 109, 100, 112, 105, 118,
               108, 113, 112, 127]

# yi do modelo
yi_modelo = [80.5 + 0.9 * x for x in xi]

# Clculo dos resduos
residuos = [y - y_hat for y, y_hat in zip(y_observado, yi_modelo)]

# Plot da distribuio dos resduos
plt.figure(figsize=(8, 5))
sns.histplot(residuos, kde=True, color='purple', bins=5)
plt.axvline(0, color='black', linestyle='--', linewidth=1)
plt.title('Distribuio dos Erros (Resduos)')
plt.xlabel('Erro (y observado - yi previsto)')
plt.ylabel('Frequncia')
plt.grid(True)
plt.tight_layout()
plt.show()
```
